```python
%matplotlib notebook
from utils.plot_functions import plot_s_animation, plot_s_snapshots

import numpy as np
import time
import pickle
import os
import matplotlib.pyplot as plt
from copy import deepcopy
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.optimize import differential_evolution

from utils.env_evaluate_functions import eval_actions
```

```python
seed=1
case = 'case_1'
opt_dir = './'+case
os.makedirs(opt_dir, exist_ok=True)
```

```python
with open('../envs_params/env_data_v1/env_eval_dict.pkl', 'rb') as input:
    envs = pickle.load(input)
rl_indices = list(range(16))
```

```python
fine_env = envs[5]
k_list = fine_env.ressim_params.k_list
envs = []
for k in k_list:
    env_ = deepcopy(fine_env)
    env_.set_k(np.array([k]))
    envs.append(env_)
```

```python
def objective_function(q_arg, *args ):
    env = args[0]
    assert q_arg.size == env.ressim_params.terminal_step*env.action_space.shape[0], 'invalid argument size'
    q_arg = q_arg.reshape(env.ressim_params.terminal_step, -1)
    _,_, rewards = eval_actions(env, q_arg)
    return -sum(rewards)
```

```python
pop_size=32
init_arg = np.random.uniform(low=env_.action_space.low[0], high=env_.action_space.high[0], size=(pop_size,env_.ressim_params.terminal_step*env_.action_space.shape[0]))
for i in rl_indices:
    env = envs[i]
    args = (env,)
    base_action = np.ones((env.ressim_params.terminal_step,env.action_space.shape[0]))
    _,_, rewards = eval_actions(env, base_action)
    print(f'case: {i} (base reward: {-round(sum(rewards), 2)})')
    bounds =  [(env.action_space.low[0],env.action_space.high[0])]*(env.ressim_params.terminal_step*env.action_space.shape[0])
    result = differential_evolution(objective_function, bounds, args=args,
                                    maxiter=1500, popsize=0.3, tol=1e-4,
                                    disp=True, polish=False,
                                    recombination=0.8, mutation=(0.5,1),
                                    init=init_arg,
                                    updating='deferred', workers=64, seed=seed)
    init_arg[0,:] = result.x
    np.save(opt_dir+'/ck_argmax_'+str(i)+'.npy', result.x)
    np.save(opt_dir+'/ck_max_'+str(i)+'.npy', result.fun)
    print(result.fun, result.x)
```
